# Descriptive Analysis

##### 1. What is the average price of Airbnb listings in the dataset?

```python
avg_price = df.price.mean()
print(f'The average price of Airbnb listings is ¥ {round(avg_price,2):,}')
```

As for each category:

```python
room_type_price = df.groupby(['room_type'])['price'].mean().round(2)
room_type_price
```

```python
plt.bar(height = room_type_price,x = room_type_price.index,color = 'purple');
plt.title('Average price of listing per room type');
```

```python
room_type_average = df.groupby(['room_type'])['price'].mean().round(2).idxmax()
print(f' {room_type_average} has the highest average price among the room types')
```

##### 2. What is the most common room type among Airbnb listings?

```python
most_com_room_typ = df.groupby(['room_type']).size().sort_values(ascending = False)
most_com_room_type = df.groupby(['room_type']).size().sort_values(ascending = False).idxmax()
print(f"The most common room type is {most_com_room_type}")
```

```python
plt.barh(width = most_com_room_typ,y = most_com_room_typ.index,color = 'darkgreen');
plt.title('Count per room type');
plt.ylabel('Room Type');
plt.xlabel('Count');
```

##### 3. What is the average number of reviews per month overall listings?

```python
print(f"The average number of reviews per month of for listing with reviews is {df.reviews_per_month.mean().round(2)}")
```

```python
avg_rating = df.groupby(['room_type']).reviews_per_month.mean().round(2).idxmax() 
avg_rating
```

```python
print(f"{avg_rating} has the highest average rating")
```

```python
sns.histplot(df.reviews_per_month,bins = 15, kde = True);
plt.title('Reviews per month Distribution')
```

##### 2. Is there a correlation between latitude and longitude with price?

```python
df[['price','latitude','longitude']].corr(method= 'pearson')
```

```python
sns.heatmap(df[['price','latitude','longitude']].corr(method= 'pearson'),annot = True);
plt.title("Pearson Correlation between price and location");
```

```python
df[['price','latitude','longitude']].corr(method= 'kendall');
```

```python
sns.heatmap(df[['price','latitude','longitude']].corr(method= 'kendall'),annot = True);
plt.title("Kendall Correlation between price and location");
```

```python
df[['price','latitude','longitude']].corr(method= 'spearman')
```

```python
sns.heatmap(df[['price','latitude','longitude']].corr(method= 'spearman'),annot = True);
plt.title("Spearman Correlation between price and location");
```

```python
sns.scatterplot(x = 'longitude',y = 'latitude', hue = 'price',data = df,palette='coolwarm',alpha = .5);
plt.title("Location vs Price");
```

# Time Series Analysis

##### 1. Has the availability of listings changed over time?If so, how?

```python
df['last_review'] = pd.to_datetime(df['last_review'])
ok = df.groupby(df['last_review'].dt.to_period('M'))['availability_365'].mean()
ok.plot(kind = 'line',linestyle = '-',marker = 'o');
plt.xlabel('Date');
plt.ylabel('Number of Listings');
plt.title('Has the availability of listings changed over time?'); 
```

After converting the last review to date and time we can see the listings took a downward trend however is slowing trying to reach the levels before 2019.

##### 2. Are the pricing affected by seasons?If so, how?

```python
price_over_time = df.groupby(df['last_review'].dt.to_period("M"))['price'].mean()
price_over_time.plot(kind = 'line',linestyle = '-',marker = 'o');
plt.xlabel('Time');
plt.ylabel('Price');
plt.title('Has the Price of listings changed over time?'); 
```

```python
df['season'] = df.last_review.dt.month.apply(lambda x:'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df.groupby(['season']).price.mean().sort_values(ascending = False).plot(kind = 'bar',color = 'royalblue');
plt.xlabel('Seasons');
plt.ylabel('Average Price');
plt.title('Are the pricing affected by seasons?');
plt.xticks(rotation = 0);
```
